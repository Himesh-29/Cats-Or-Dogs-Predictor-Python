## Making the model and then training it

```python
import pickle
features=pickle.load(open('features.pkl','rb'))
labels=pickle.load(open('labels.pkl','rb'))
```

```python
#Feature scaling
features=features/255
```

```python
features
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
model=Sequential()

model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(128,input_shape=features.shape[1:],activation="relu"))

#Output layer
model.add(Dense(2,activation="softmax"))
```

## Compiling the model

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

## Fitting the dataset in the model

```python
(x_train,x_test,y_train,y_test)=train_test_split(features,labels,test_size=0.2)
```

```python
x_train.shape
```

```python
model.fit(x_train,y_train,epochs=10,validation_split=0.1)
```

```python
model.summary()
```

```python
prediction=model.predict(x_test)
prediction=np.argmax(prediction,axis=1)
```

```python
print("Accuracy score: "+"{:.2f}".format(accuracy_score(y_test,prediction)*100)+"%")
```

```python
model.save('catsAndDogs.h5')
```
